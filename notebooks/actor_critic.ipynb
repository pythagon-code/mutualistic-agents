```python
from fnn import FNN
from modular_network import ModularNetwork
import torch
from torch import nn, optim

class ActorCritic(nn.Module):
    def __init__(self, state_dim, action_dim, hidden_dim):
        super().__init__()

        self._state_dim = state_dim
        self._action_dim = action_dim
        self._hidden_dim = hidden_dim

        self._actor = ModularNetwork(
            input_dim = state_dim,
            embed_dim = hidden_dim,
            num_hidden_layers = 6,
            num_encoders = 2,
            tree_depth = 2,
            fan_in = 2,
            output_dim = action_dim,
            output_activation = nn.Tanh(),
        )

        self._critic = FNN(
            input_dim = hidden_dim + action_dim,
            embed_dim = hidden_dim,
            num_hidden_layers = 6,
            num_encoders = 2,
            tree_depth = 2,
            fan_in = 2,
            output_dim = 1,
        )

    def actor_forward(self, state):
        return self._actor(state)

    def critic_forward(self, hidden, action):
        return self._critic(torch.cat([hidden, action], dim = 1))
```
